```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("data100All.csv")
df[df.columns[1]] = df[df.columns[1]] / 10 ** 9
df[df.columns[2]] = df[df.columns[2]] / 10 ** 9

print(df)

print(df[df.columns[-1]])
```

```python
labels = ("MNK", "MKN", "NMK", "NKM", "KMN", "KNM", "DGEMM")
width = 0.35
x = np.arange(7)

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, df[df.columns[1]], width, label="L1 D-cache hits")
rects2 = ax.bar(x + width/2, df[df.columns[2]], width, label="L1 D-cache misses")
ax.set_title("Cache hit/miss comparison of different n k m orders")
ax.set_ylabel("Number of hits/misses (billions)")
ax.set_xticks(x, labels)
ax.legend()
plt.savefig("Cache HitMiss.jpg")
plt.show()


```

```python
plt.bar(df[df.columns[-1]], df[df.columns[0]])
plt.title("Runtime comparison of different n k m orders")
plt.xticks(np.arange(7), labels=labels)
plt.ylabel("Runtime (s)")
plt.savefig("Runtimes.jpg")
plt.show()
```
